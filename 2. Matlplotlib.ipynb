# Matplotlib

## Line Plot

```python
import numpy as np
from matplotlib import pyplot as plt 
```

```python
# make an array using arange func()
# create x
x = np.arange(1,11)
x
```

```python
# create y which is function of x
y = 3*x
y
```

```python
# plot 

plt.plot(x,y, color = 'g', linestyle= '-', linewidth = 4)
plt.title("y = 3*x")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

### Adding two lines in same plot

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1, color = 'g', linestyle= '-', linewidth = 4)
plt.plot(x,y2, color = 'r', linestyle= ':', linewidth = 4)
plt.title("Two Line In Same Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

### Adding subplots

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1, color = 'g', linestyle= '-', linewidth = 4)
plt.title("y = 2*x")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()

plt.subplot(1,2,2)
plt.plot(x,y2, color = 'r', linestyle= ':', linewidth = 4)
plt.title("y = 3*x")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()

plt.show()
```

## Bar plot

```python
Students = {"zia": 90, "Jalil": 97, "Junaid": 98} 
Students
```

```python
name = list(Students.keys())
name
```

```python
marks = list(Students.values())
marks
```

```python
plt.bar(name, marks)
plt.show()
```

### Horizontal plot

```python
plt.barh(name, marks, color="g")
plt.grid()
plt.show()
```

### Creating a Scatter plot

```python
x = [12, 4, 5, 34, 65, 78, 2, 13, 90]
y = [3,2,8,9,1,5,6,7,4]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
b = [4,5,2,3,9,8,7,1,3]
```

```python
plt.scatter(x,y, marker="*", c = "r", s = 100)
plt.scatter(x,b, c = "g", s = 100)
plt.grid()
plt.show()
```
